# Customer Churn Strategy Recommendations

## Objective

Translate model outputs into actionable retention strategies that reduce customer churn and protect recurring revenue.

This analysis focuses on identifying high-risk customer segments, estimating business impact, and recommending targeted interventions.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

df["TotalCharges"] = df["TotalCharges"].fillna(
    df["TotalCharges"].median()
)

df["Churn"] = df["Churn"].map({
    "Yes": 1,
    "No": 0
})
```

```python
high_risk = df[
    (df["Contract"] == "Month-to-month")
    & (df["tenure"] < 12)
]

high_risk.shape
```

```python
high_risk["MonthlyCharges"].describe()
```

```python
high_risk["Churn"].mean()
```

```python
revenue_at_risk = df[df["Churn"] == 1]["MonthlyCharges"].sum()

print(f"Monthly revenue lost from churned customers: ${revenue_at_risk:,.2f}")
```

```python
annualized_risk = revenue_at_risk * 12

print(f"Estimated annualized revenue at risk: ${annualized_risk:,.2f}")
```

## Recommended Actions

### 1. Prioritize Month-to-Month Customers

The analysis consistently identifies month-to-month subscribers as the highest-risk group. Retention campaigns should focus on converting these customers to longer-term contracts.

### 2. Improve First-Year Customer Retention

Customers with low tenure exhibit elevated churn rates. Investment in onboarding, customer success outreach, and early engagement programs could reduce churn during the most vulnerable stage of the customer lifecycle.

### 3. Monitor High-Charge Customers

Customers with higher monthly charges are more likely to leave. Personalized offers and proactive outreach may improve retention among these high-value accounts.

### 4. Use Predictive Scoring for Outreach

Rather than contacting all customers, the company should prioritize intervention efforts toward customers identified as high-risk by the predictive model.
